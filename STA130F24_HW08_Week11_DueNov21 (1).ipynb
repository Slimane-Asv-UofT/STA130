A **Classification Decision Tree** is a method used to solve problems where the goal is to assign items into categories or classes based on their features. It works by asking a series of "yes" or "no" questions about the data, creating a tree-like structure (like in probabilities) where each decision splits the data into smaller groups. Eventually, each group corresponds to a specific class.

**Classification Decision Trees** address **classification problems**, where the output is a class (category) rather than a numerical value.

I.E. : Medical Diagnosis: Classifying a tumor as "Benign" or "Malignant" based on patient data (e.g., size, texture, etc.).

A **Classification Decision Tree** predicts by splitting data based on conditions at each level of the tree.
I.E. : 
If you're classifying animals:
Does it have fur? (Yes/No)
Does it lay eggs? (Yes/No)
Is it warm-blooded? (Yes/No)
Each branch narrows down the possibilities until the tree reaches a final decision (a class).

How Multiple Linear Regression Makes Predictions

In contrast, Multiple Linear Regression predicts numerical values rather than categories. It uses a mathematical formula to combine multiple features (e.g., y = a*x1 + b*x2 + c), where each feature contributes proportionally to the outcome.

### 2. Continue your ChatBot session and explore with your ChatBot what real-world application scenario(s) might be most appropriately addressed by each of the following *metrics* below: provide your answers and, in your own words, *concisely explain your rationale for your answers.*<br>


1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

**Accuracy**
Example: A facial recognition system at an airport security checkpoint.
Details: The system needs to correctly identify passengers (true positives) while not misidentifying others (true negatives).
Why Accuracy ? Because both false positives (wrongly identifying an innocent person as a threat, FP in the formula above) and false negatives (failing to detect a threat, FN in the formul above) need to be minimized equally, so the important factor is overall accuracy.

**Sensitivity**
Example: Breast cancer screening using mammograms.
Details: The test identifies individuals with cancer (true positives) and those without it (false negatives).
Why Sensitivity ? Because it's better to be safe than sorry, so minimizing the FN is way more important than the FP.

**Specificity**
Example: Fraud detection for credit card transactions.
Details: The system identifies fraudulent transactions (true negatives) and misclassifies legitimate transactions as fraud (false positives).
Why Specificity ? Because a high false positive rate (flagging normal transactions as fraud) can lead to customer dissatisfaction and blocked accounts, so specificity is prioritized to ensure that legitimate transactions are rarely flagged incorrectly.

**Precision**
Example: Product recommendations on an e-commerce site (e.g., Amazon).
Details: The system suggests products that might interst the user (true positives) and occasionally shows irrelevant ones (false positives).
Why Precision ? Showing irrelevant products will not lead to a sale, and therefore lower the value of the ad.

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

# Define and train clf (first model)
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X_train, y_train)

# Define and train clf2 (second model with a different max_depth)
clf2 = DecisionTreeClassifier(max_depth=3, random_state=42)
clf2.fit(X_train, y_train)

# Predict with clf and clf2
y_pred_clf = clf.predict(X_test)
y_pred_clf2 = clf2.predict(X_test)

# Compute confusion matrices
cm_clf = confusion_matrix(y_test, y_pred_clf)
cm_clf2 = confusion_matrix(y_test, y_pred_clf2)

# Define a function to calculate metrics
def calculate_metrics(cm):
    TN, FP, FN, TP = cm.ravel()
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    sensitivity = TP / (TP + FN) if (TP + FN) != 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) != 0 else 0
    return accuracy, sensitivity, specificity

# Calculate metrics for clf
accuracy_clf, sensitivity_clf, specificity_clf = calculate_metrics(cm_clf)

# Calculate metrics for clf2
accuracy_clf2, sensitivity_clf2, specificity_clf2 = calculate_metrics(cm_clf2)

# Print results
print("Confusion Matrix for clf:")
print(cm_clf)
print(f"Metrics for clf:\nAccuracy: {accuracy_clf:.2f}, Sensitivity: {sensitivity_clf:.2f}, Specificity: {specificity_clf:.2f}")

print("\nConfusion Matrix for clf2:")
print(cm_clf2)
print(f"Metrics for clf2:\nAccuracy: {accuracy_clf2:.2f}, Sensitivity: {sensitivity_clf2:.2f}, Specificity: {specificity_clf2:.2f}")

# Optional: Visualize the trees
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)
plt.title("Decision Tree: clf")
plt.show()

plt.figure(figsize=(10, 6))
plot_tree(clf2, feature_names=['List Price'], class_names=['Paperback', 'Hardcover'], filled=True)
plt.title("Decision Tree: clf2")
plt.show()
````
