I used ChatGTP to make the code below (especially for the visual representation), but I annotated it myself afterwards and tried to redo the boothstrap part to be sure I understanded it. The goal of this code is to calculate the mean of final health scores, and then use boothstrap to verify the interval of confidence and check if the mean is (likely) representative. Then, if the mean is representative, check the difference with the mean of the initial test score, and compare. Higher the difference, more effective is the vaccin. 

````python
import numpy as np
import matplotlib.pyplot as plt

# Data: Initial and Final HealthScores of patients
initial_health_scores = np.array([84, 78, 83, 81, 81, 80, 79, 85, 76, 83])
final_health_scores = np.array([86, 86, 80, 86, 84, 86, 86, 82, 83, 84])

# Calculate the means of the samples
mean_initial = np.mean(initial_health_scores)
mean_final = np.mean(final_health_scores)
print("This is the sample mean of the initial health score: ", mean_initial)
print("This is the sample mean of the health score after the vaccin: ", mean_final)

# Calculate the mean difference
mean_difference = mean_final - mean_initial
print("This is the difference of the means: ", mean_difference)

# Set the random seed for reproducibility
np.random.seed(42)

# Number of bootstrap samples
n_bootstraps = 1000

# Arrays to store bootstrap means
bootstrap_means_final = np.zeros(n_bootstraps)
bootstrap_means_initial = np.zeros(n_bootstraps)
bootstrap_mean_differences = np.zeros(n_bootstraps)

# Perform bootstrap resampling
for i in range(n_bootstraps):
    # Resample with replacement from the original data
    bootstrap_sample_final = np.random.choice(final_health_scores, size=len(final_health_scores), replace=True)
    bootstrap_sample_initial = np.random.choice(initial_health_scores, size=len(initial_health_scores), replace=True)
    
    # Compute the means of the resampled data
    bootstrap_means_final[i] = np.mean(bootstrap_sample_final)
    bootstrap_means_initial[i] = np.mean(bootstrap_sample_initial)
    bootstrap_mean_differences[i] = bootstrap_means_final[i] - bootstrap_means_initial[i]

# Compute standard deviations of the bootstrap means
std_bootstrap_means_final = np.std(bootstrap_means_final)
std_bootstrap_means_initial = np.std(bootstrap_means_initial)
std_bootstrap_mean_difference = np.std(bootstrap_mean_differences)

# Plot histograms of the bootstrap means
plt.figure(figsize=(12, 6))

# Histogram for FinalHealthScore Means
plt.subplot(1, 2, 1)
plt.hist(bootstrap_means_final, bins=20, edgecolor='k', alpha=0.7)
plt.title('Bootstrap Distribution of Mean FinalHealthScore')
plt.xlabel('Mean FinalHealthScore')
plt.ylabel('Frequency')
plt.axvline(mean_final, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean_final:.2f}')
plt.axvline(mean_final + std_bootstrap_means_final, color='green', linestyle='dotted', linewidth=2, label=f'+1 Std Dev ({mean_final + std_bootstrap_means_final:.2f})')
plt.axvline(mean_final - std_bootstrap_means_final, color='green', linestyle='dotted', linewidth=2, label=f'-1 Std Dev ({mean_final - std_bootstrap_means_final:.2f})')
plt.legend()

# Histogram for Mean Differences
plt.subplot(1, 2, 2)
plt.hist(bootstrap_mean_differences, bins=20, edgecolor='k', alpha=0.7, color='orange')
plt.title('Bootstrap Distribution of Mean Difference (Final - Initial)')
plt.xlabel('Mean Difference')
plt.ylabel('Frequency')
plt.axvline(mean_difference, color='red', linestyle='dashed', linewidth=2, label=f'Mean Difference = {mean_difference:.2f}')
plt.axvline(mean_difference + std_bootstrap_mean_difference, color='green', linestyle='dotted', linewidth=2, label=f'+1 Std Dev ({mean_difference + std_bootstrap_mean_difference:.2f})')
plt.axvline(mean_difference - std_bootstrap_mean_difference, color='green', linestyle='dotted', linewidth=2, label=f'-1 Std Dev ({mean_difference - std_bootstrap_mean_difference:.2f})')
plt.legend()

plt.tight_layout()
plt.show()
````

The null hypothese is the idea that the vaccin has no effect, and the variation in the means would just be due to the fact that the sample is not representative of the population.

Firstly, we calculate the mean of the sample before the vaccin, and then after.
Secondly, we calculate the difference between the initial mean and the sample mean.
Thirdly, we boothstrap the sample "after the vaccin" to be sure the sample mean is in the are of confidence.
Here, we only need one standart deviation right and left, so it really likely is.
Fourthly, we create a histogramm for visual representation (the blue one).
Finally, we create a histgramm for the mean difference.

After the boothstrap, we see that the mean difference is quite higher than 0, therefore we can conclude that the vaccin is indeed working.
